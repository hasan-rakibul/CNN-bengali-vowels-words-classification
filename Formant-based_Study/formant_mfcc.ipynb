```python
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
mfcc_path="/media/rakibul/WorkFolder/Work/Thesis/Speech_Recognition/Bangla_Vowel_Recognition/Using MFFC/Result/"
formant_path="./Result/"
```

```python
train=[]
for trial in os.listdir(mfcc_path):
    if len(train)==0:
        train=np.load(mfcc_path+trial)
    else:
        train=np.vstack((train,np.load(mfcc_path+trial)))
```

```python
train
```

```python
loss=train[:,0]
acc=train[:,1]
```

```python
n_run=np.arange(1,8)
plt.figure(figsize=(10,4))
plt.plot(n_run,loss,'bo',label='Training Loss')
plt.plot(n_run,acc,'b',label='Training Accuracy')
plt.xlabel('Number of Data/Class')
plt.ylabel('Loss or Accuracy')
plt.legend()
plt.savefig('./Result/Loss_Accuracy.png',format='png',dpi=1200)
```

```python
for letter in range(1,10):
    variation=[]
    for i in range(1,8):
        filename=formant_path+"COV_trial"+str(i)+"_vowel_"+str(letter)+".npy"
        
        if len(variation)==0:
            variation=np.load(filename)
        else:
            variation=np.vstack((variation,np.load(filename)))
    
    plt.figure(figsize=(10,4))
    plt.plot(n_run,variation[:,0],'bo')
    plt.xlabel('Number of Data/Class')
    plt.ylabel('Coefficient of variation')
    plt.savefig('./Result/Variation'+str(letter)+'.png',format='png',dpi=1200)
```
