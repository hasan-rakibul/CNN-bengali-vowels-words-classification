```python
url ='/media/rakibul/WorkFolder/Work/Thesis/Speech_Recognition/Bangla_Vowel_Recognition/Using_Formant/data/Formant1/'
name ='vowel1_0_Formant_frq.txt'
file = url+ name
data_read = pd.read_csv(file, delim_whitespace=True,header=None,dtype=np.float64,na_values="--undefined--")
```

```python
data_read.describe
```

```python
scaler=MinMaxScaler()
scaler.fit(data_read)
formants=scaler.transform(data_read)
```

```python
formants
```

```python
max_len =150
if (max_len > formants.shape[0]):
    pad_width=max_len-formants.shape[0]
    heading_zero=int(pad_width/2)
    trailing_zero=pad_width-heading_zero
    formants=np.pad(formants,pad_width=[(heading_zero,trailing_zero),(0,0)],mode='constant')
else:
    formants=formants[:max_len,:]
```
